## 3) Series Methods
**Question:**
Given `scores = pd.Series([90, 85, 88, 92, 76])`:
A. Find the mean.
B. Find the standard deviation.
C. Find the sum.

```python
import pandas as pd
scores = pd.Series([90, 85, 88, 92, 76])
# A. Mean
print(scores.mean())

# B. Standard deviation
print(scores.std())

# C. Sum
print(scores.sum())
```

## 4) Series Attributes
Using the `scores` Series above:
A. Print its `dtype`.
B. Print its `size`.
C. Check if all values are unique.

```python
# A. dtype
print(scores.dtype)

# B. size
print(scores.size)

# C. is_unique
print(scores.is_unique)
```
