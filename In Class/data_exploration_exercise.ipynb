## 6) Summarizing Data
**Question:**
A. Compute the average profit margin.
B. Find the maximum revenue and the company that earned it.
C. Group companies by `Industry` and compute their mean revenue.

```python
# A. average profit margin
print(companies.Profit_Margin.mean())

# B. max revenue and company
print(companies.Revenue.max(),companies[companies.Revenue==companies.Revenue.max()]['Company'])

# C. group by industry
print(companies.groupby(by='Industry').agg({'Revenue':'mean'}))
```
